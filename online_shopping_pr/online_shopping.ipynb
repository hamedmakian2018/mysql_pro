# Creating table named Users

```python
db_name ='online_shopping'
table_name  = 'Users'
table_columns = """
        (user_id int primary key auto_increment,
        username varchar(50) not null unique,
        password varchar(50) not null,
        name varchar(100) not null,
        email varchar(100) not null,
        contact_number varchar(30),
        street varchar(255),
        city varchar(100),
        state varchar(100),
        postal_code varchar(20),
        country varchar(100),
        registration_date timestamp default current_timestamp);

"""
qu_cr_tb = f"""CREATE TABLE IF NOT EXISTS {table_name} {table_columns}"""
cux(qu_cr_tb)
print (f'A table named {table_name} created successfully in {db_name} database ')
print ('\n ')
print ('#'*70)
print ('\n ')

# Show tables in database
qu_sh_tb = """SHOW TABLES"""
cux(qu_sh_tb)
tables = cur.fetchall()
for table in tables:
    print (table[0])

print ('\n ')
print ('#'*70)
print ('\n ')

# Show coulmns of table
print (f'columns in  {table_name}:')
qu_col_tb = f"""DESCRIBE {table_name}"""
cux(qu_col_tb)
columns = cur.fetchall()
for column in columns:
    print (column)

```
